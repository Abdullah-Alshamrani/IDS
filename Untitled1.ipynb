```python
#Import Pandas
import pandas as pd
#import matplot
import matplotlib.pyplot as plt
```

```python
#Load the dataset
url = "https://raw.githubusercontent.com/bforoura/IDS/main/HW4/adult_training.csv"
data = pd.read_csv(url)
```

```python
# Workclass and sex contingency table Plor
workclass_sex_contingency = pd.crosstab(data['workclass'], data['sex'])
print(workclass_sex_contingency)
workclass_sex_contingency.plot(kind='bar', stacked=True)
plt.title(" Workclass vs. Sex")
plt.xlabel("Workclass")
plt.ylabel("Count")
plt.show()
```

```python
# contingency table for marital status/sex, and plot it
sex_marital_contingency = pd.crosstab(data['sex'], data['marital-status'])
print(sex_marital_contingency)
sex_marital_contingency.plot(kind='bar', stacked=True)
plt.title("Sex vs. Marital Status")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()

```

```python
#Create a new data frame, adultOver40, for those whose age > 40.
over40 = data[data['age'] > 40]
```

```python
#Recreate the contingency table of sex and marital status for adultOver40.
sex_marital_contingency_over40 = pd.crosstab(over40['sex'], over40['marital-status'])
print(sex_marital_contingency_over40)
```

```python
#Determine whether any outliers exist for the education field.
education_outliers = data[(data['education'].isna()) | (data['education'] == ' ?')]
print("Outliers in education:", len(education_outliers))
```

```python
 #Standardize (zero mean) the age variable, and identify how many outliers there are. What is the most extreme outlier?
mean_age = data['age'].mean()
std_age = data['age'].std()
data['standardized_age'] = (data['age'] - mean_age) / std_age

z_score_threshold = 3
age_outliers = data[(data['standardized_age'] > z_score_threshold) | (data['standardized_age'] < -z_score_threshold)]
print("Number of age outliers:", len(age_outliers))
print("Most extreme age outlier:")
print(age_outliers[['age', 'standardized_age']].max())
```

```python
#Age anomaly? Select only records with age  at least 80, and construct a histogram of age. Explain what you see.
age_over_80 = data[data['age'] >= 80]
age_over_80['age'].hist()
plt.title("Histogram of Age (Age >= 80)")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()


# more 90 people live? but less from 80 to 88?
```
